```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
feature_stack = np.load('../data/processed/feature_stack.npy')
labels = np.load('../data/processed/labels.npy')
```

```python
print(f"Feature Stack Shape: {feature_stack.shape}") 
print(f"Labels Shape: {labels.shape}")
```

```python
fig, axes = plt.subplots(1, 4, figsize=(20, 5))
```

```python
titles = ['Elevation (DEM)', 'Fuel (Bhuvan LULC)', 'Proximity (Roads)', 'Ground Truth (Fires)']
data = [feature_stack[0], feature_stack[1], feature_stack[2], labels]
```

```python
for i, ax in enumerate(axes):
    im = ax.imshow(data[i], cmap='magma' if i != 1 else 'terrain')
    ax.set_title(titles[i])
    plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
```

```python
plt.tight_layout()
plt.show()
```

```python
fire_indices = np.where(labels== 1)
if len(fire_indices[0]) >0:
    print("Average Elevation of Fire sites:",np.mean(feature_stack[0][fire_indices]))
    print("Most common Fuel code at Fire sites:",np.bincount(feature_stack[1][fire_indices].astype(int)).argmax())
```
